#### Получение Z-оценки по P-значению

https://planetcalc.ru/7803/

```python
def z_stat (x, mu, s, n):
    return (x-mu)/(s/math.sqrt(n))
```

```python
z_stat(13.5,14,2,64)
```

```python
print(norm.cdf(-2.0))
```

```python
array = np.array([6.1,9.2,11.5,8.6,12.1,3.9,8.4,10.1,9.4,8.9])
```

```python
t.cdf(z_stat(array.mean(),10,array.std(ddof=1),len(array)), len(array) - 1)
```

```python
def p_stat (p,x,n):
    return (p-x)/(math.sqrt((x*(1-x)/n)))
```

```python
p = 87/125
p_stat(p, 0.75, 125)
```

```python
norm.cdf(p_stat(p, 0.75, 125))
```
