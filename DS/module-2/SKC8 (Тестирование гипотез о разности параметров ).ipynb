```python
def p_stat (p_one,p_two,n_one,n_two):
    p = (p_one*n_one+p_two*n_two)/(n_one+n_two)
    return (p_one-p_two)/(math.sqrt((p*(1-p)*((1/n_one)+(1/n_two)))))
```

```python
p_one = 62/100
p_two = 29/100
n_one = 100
n_two = 100
z = p_stat(p_one,p_two,n_one,n_two)
p_value = norm.cdf(z)
```

```python
print(z, p_value)
```
