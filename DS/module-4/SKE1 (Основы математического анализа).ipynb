# Что такое функция?<a name="1"></a>

*Определение*: __Функцией__ в математике называется правило, по которому можно сопоставить элементы двух множеств.

*Формальное определение*: Для любого элемента *__х__* из множества *__Х__* существует единственный элемент *__y__* из множества *__Y__*, такой что __f(x) = y__:

Проиллюстрировав на картинке множество *__Х__* и множество *__Y__*, видим: у каждого элемента *__x__* из *__X__* должна быть пара *__y__* в *__Y__*. Таким образом, под действием функции *__f__* все множество *__X__* переходит в некоторую часть множества *__Y__*.

Восклицательный знак в формальном определении функции говорит о том, что такая пара *__x – y__* должна быть одна. Это значит, что нельзя одному *__х__* поставить в соответствие два разных *__y__*. И в то же время двум *__х__* — один *__y__*: сколько угодно!

Обозначение:

Функции одной переменной часто изображают графически в декартовой системе координат

# Производная. Общие понятия<a name="4"></a>

Если функция достаточно «хорошая» рядом с известной нам точкой, то к ней можно привести касательную. И если не слишком далеко отойти от x0, то график функции будет очень близок к касательной.

Другими словами, в небольшой окрестности x0 наша функция будет почти прямой. Это значит, что её приращение можно считать как приращение линейной функции. 

* Синяя прямая — это __касательная__ к графику в точке *__A__*.  

* *__α__* — угол наклона касательной.  

* Точка *__A__* на графике имеет координаты *__(x0,f(x0))__* то есть *__(1,1).__*  

* Красной точкой обозначено настоящее значение нашей функции *__y1__* в точке *__x1__*.  

* Точка *__B__* на касательной — та точка, куда мы попали бы, если бы вышли из *__A__* не по графику, а по касательной.  

* *__b__* — линейное приращение.  

* Треугольник *__ABC__* прямоугольный, а значит *__b__* можно вычислить как *__Δx·tgα__*.  

* А значит, нужное нам значение *__y1__* приближённо равно *__y0+Δx⋅tgα__*.

Осталось понять, как вычислить *__tgα__*.

* Для этого давайте нарисуем хорду и обозначим за *__Δy__* настоящее приращении функции *__f__*.
* С уменьшением *__Δx__* хорда будет все больше похожа на касательную, а её угол наклона — на угол *__α__*.
* Если устремить *__Δx__* к нулю, то хорда превратится к касательную, а отношение *__Δy__* к *__Δx__* — в *__tgα__*.

*Определение*:  __Производная__ — есть предел приращений и тангенс угла наклона касательной.

### Линеаризация

*Определение*: Задача о приближении, с помощью которой мы только что ввели понятие производной, называется __линеаризацией__.

У нас есть значение функции в точке *__x0__*. А хотим мы приближенно узнать ее значение в точке *__x1__*.

Мы показали, что если *__x1__* находится не очень далеко от *__x0__*, то отношение приращений примерно равно производной в точке *__x0__*.

Чем ближе новая точка расположена к старой, тем точнее приближение.

На идее линеаризации основываются *marginal effects* в нелинейных регрессиях.
